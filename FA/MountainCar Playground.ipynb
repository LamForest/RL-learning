```python
%matplotlib inline

import gym
import numpy as np
from matplotlib import pyplot as plt
```

```python
env = gym.envs.make("MountainCar-v0")
```

```python
env.reset()
plt.figure()
plt.imshow(env.render(mode='rgb_array'))

# [ ]
for x in range(1000):
    
    a = np.random.choice([0,1,2])
    
    env.step(a)
    
    if x % 100 == 0:
        plt.figure()
        plt.imshow(env.render(mode='rgb_array'))

env.close()
```
